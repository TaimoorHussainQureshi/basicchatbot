```python
from transformers import pipeline
from fpdf import FPDF
```

```python
chatbot = pipeline("conversational", model="microsoft/DialoGPT-medium")
```

```python
def chat_and_generate_pdf():
    conversation_history = []
    print("Chatbot: Hi! How can I help you today? (type 'quit' to stop)")

    while True:
        user_input = input("You: ")
        if user_input.lower() == 'quit':
            break

        response = chatbot(user_input)[0]
        conversation_history.append(f"You: {user_input}")
        conversation_history.append(f"Chatbot: {response['generated_text']}")
        print(f"Chatbot: {response['generated_text']}")

    return conversation_history
```

```python
def generate_pdf(conversation):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)

    for line in conversation:
        pdf.cell(200, 10, txt=line, ln=True)

    pdf.output("/content/conversation.pdf")
    print("PDF generated: conversation.pdf")
```

```python
from transformers import Conversation, pipeline

def chat_and_generate_pdf():
    chatbot = pipeline("conversational", model="microsoft/DialoGPT-medium")
    conversation_history = []
    print("Chatbot: Hi! How can I help you today? (type 'quit' to stop)")
    conversation = Conversation()

    while True:
        user_input = input("You: ")
        if user_input.lower() == 'quit':
            break

        conversation.add_user_input(user_input)
        response = chatbot(conversation)
        response_text = response.generated_responses[-1]
        conversation_history.append(f"You: {user_input}")
        conversation_history.append(f"Chatbot: {response_text}")
        print(f"Chatbot: {response_text}")

    return conversation_history
```

```python
conversation = chat_and_generate_pdf()
generate_pdf(conversation)

```
